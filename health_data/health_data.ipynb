```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import altair as alt
```

```python
#reading excel spreadsheet
health_data = pd.read_csv('/content/drive/MyDrive/Year V/Data Visualization/health-income-withregions (1).csv')
health_data.head(5)
```

```python
# 1.1 - Scatter Plot 3 Attributes - Country/Region as a colour
alt.Chart(health_data).mark_circle().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    color=alt.Color('region', scale=alt.Scale(scheme='tableau20')),
    # size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    # size='population',
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# 1.2 - Scatter Plot 3 Attributes - Country/Region as a marker
alt.Chart(health_data).mark_point().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    shape = 'region',
    # size=alt.Size('region', scale=alt.Scale(range=[10, 100])),    # size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# 1.3 - Scatter Plot 3 Attributes - Country/Region as size
alt.Chart(health_data).mark_circle().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    # shape = 'region',
    size=alt.Size('region', scale=alt.Scale(range=[1, 100])),    # size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# 2.1 - Scatter Plot 4 Attributes - Region as Marker, population as Size
alt.Chart(health_data).mark_point().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    # color=alt.Color('region', scale=alt.Scale(scheme='tableau20')),
    shape='region',
    size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    # size='population',
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# 2.2 - Scatter Plot 4 Attributes - Region as Marker, population as brightness
alt.Chart(health_data).mark_point().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    shape='region',
    # color=alt.Color('region', scale=alt.Scale(scheme='tableau20')),
    # size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    # size=alt.value('population:Q', bin=True),
    opacity='population',
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# 2.3 - Scatter Plot 4 Attributes - Region as Colour, population as Colour
alt.Chart(health_data).mark_point().encode(
    alt.X('health', scale=alt.Scale(zero=False)),
    alt.Y('income', scale=alt.Scale(zero=False, padding=1)),
    color=alt.Color('population', scale=alt.Scale(scheme='turbo')),
    # size=alt.Size('population', scale=alt.Scale(range=[10, 1000])),
    # shape=alt.value('population:Q', bin=True),
    shape='region',
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)
```

```python
# Alt - Stacked Bar Chart
bars = alt.Chart(health_data).mark_bar().encode(
    x=alt.X('health:O', bin=True),
    y=alt.Y('population:Q', stack='zero'),
    color=alt.Color('country', scale=alt.Scale(scheme='tableau20')),
    opacity='income',
    tooltip=['country', 'region', 'health', 'income', 'population']
).interactive().properties(width=1300,height=350)

bars
```
